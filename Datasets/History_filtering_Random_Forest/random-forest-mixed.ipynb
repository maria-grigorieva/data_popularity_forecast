```python
import pandas as pd
import pprint

import matplotlib
# for interactive plots in Jupyter:
# matplotlib.use('qtagg')

from random_forest_train_test_validate import train_test_validate
```

```python
history_lens = [
    pd.Timedelta('365 days'),
    pd.Timedelta('270 days'),
    pd.Timedelta('180 days'),
    pd.Timedelta('90 days'),
]

span_thresholds = [
    pd.Timedelta('90 days'),
    pd.Timedelta('60 days'),
    pd.Timedelta('40 days'),
    pd.Timedelta('20 days'),
]

# history_len = pd.Timedelta('365 days')
# span_threshold = pd.Timedelta('90 days')

accesses_threshold = 30
source_csv = (
    '/home/mshubin/Desktop/datasets-popularity-task-data/' +
    'datasets_popularity_DAOD_mixed_' + 
    'higgd-topq-susy-phys_mc16_13TeV_202308141254.csv'
)
finish_date_train_test = pd.Timestamp('2023-03-21')
finish_date_validate = pd.Timestamp('2023-04-06')
horizon = pd.Timedelta('7 days')
Ntest = 3000
```

```python
results_list = []

for history_len, span_threshold in zip(history_lens, span_thresholds):
    print(f'\nRUNNING TEST FOR HISTORY LEN = {history_len.days} DAYS\n')
    res = train_test_validate(
        history_len = history_len,
        span_threshold = span_threshold,
        accesses_threshold = accesses_threshold,
        source_csv = source_csv,
        finish_date_train_test = finish_date_train_test,
        finish_date_validate = finish_date_validate,
        horizon = horizon,
        Ntest = Ntest,
    )
    results_list.append((history_len, res))
```

## Print results

```python
for hist_len, res in results_list:
    print(hist_len.days, 'days')
    print('   ', 'accuracy (test):', res['test_accuracy'])
    print('   ', 'accuracy (val):', res['val_accuracy'])

print()

for hist_len, res in results_list:
    print()
    print(hist_len.days, 'days')
    print()
    print(res['test_confusion_matrix'])
    print(res['test_classification_report'])
    print(res['val_confusion_matrix'])
    print(res['val_classification_report'])
```
